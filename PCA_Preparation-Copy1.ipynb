```python
ds1 = (new_df.sort_values(['Indicator', 'Period'], ascending = False)
         .drop_duplicates(['Indicator', 'Location'], keep='first'))

#ds1.sort_values(['Indicator', 'Period'], ascending = False)
ds1.shape
```

```python
ds1.sample(50)
```

```python
ds1 =ds1[ds1['First Tooltip'] != 'No data']
```

```python
word = '<'
ds1 = ds1[~ds1["First Tooltip"].str.contains(word, na=False)]

ds1.shape
```

```python
# cleaning out numbers that have values between brackets
ds1['First Tooltip'] = ds1['First Tooltip'].str.split('[', 1).str[0].str.strip()
ds1['First Tooltip'] = pd.to_numeric(ds1['First Tooltip'])
```

```python
ds1.info()
```

```python
ds1.nunique()
```

```python
ds1['% OF HALE in life expectancy'].value_counts()
```

```python
ds2 = ds1[['Location' , 'Period' , 'Indicator' , 'First Tooltip']]
```

```python
ds3 = ds2[ds2['First Tooltip'].notna()]
ds3.nunique()
```

```python
ds4 = ds3.dropna(how='any')
```

```python
ds4.nunique()
```

```python
ds5 = ds4.pivot_table(columns='Indicator', index='Location', values='First Tooltip').reset_index()
ds5
```

```python
((ds5.isnull().sum()/ds5.isnull().count()) * 100).sort_values(ascending=False)
```

```python
ds6 = ds5.drop(columns=['Population using safely managed sanitation services (%)' , 
                        'Population with basic handwashing facilities at home (%)',
                       'Malaria incidence (per 1 000 population at risk)',
                       'New HIV infections (per 1000 uninfected population)',
                       'Married or in-union women of reproductive age who have their need for family planning satisfied with modern methods (%)',
                       ])

((ds6.isnull().sum()/ds6.isnull().count()) * 100).sort_values(ascending=False)
```

```python
ds7 = ds6.dropna(axis = 0, how='any')
    
ds7.info()
```

```python
ds7.to_csv('WHO Data for PCA.csv', index=False)
```
