```python
import os
import pandas as pd
import glob
writer = pd.ExcelWriter('out.xlsx', engine='xlsxwriter')
path = "C:/Users/motta/OneDrive - IE Students/IE/Data Viz/Assignment 2/data"
all_files = glob.glob(os.path.join(path, "*.csv"))
for f in all_files:
    df = pd.read_csv(f)
    if (len((os.path.basename(f))[:-4]) <=31):
        df.to_excel(writer, sheet_name=((os.path.basename(f))[:-4]), index=False)
    else:
        df.to_excel(writer, sheet_name=((os.path.basename(f))[:31]), index=False)
    print((os.path.basename(f))[:-4])
writer.save()
```

```python
#combine all files in the list
combined_csv = pd.concat([pd.read_csv(f) for f in all_files ])
#export to csv
combined_csv.to_csv("combined_csv.csv", index=False, encoding='utf-8-sig')
```
