```python
import os
import pandas as pd
import glob
import numpy as np

df = pd.read_excel('combined_indicators_selected_Inidcators_added_v3.xlsx', sheet_name='PCA_All_Ind_2015_v2')

df.head()
```

```python
df.sample(50)
```

```python
# cleaning out numbers that have values between brackets
df['First Tooltip'] = df['First Tooltip'].str.split('[', 1).str[0].str.strip()
df['First Tooltip'] = pd.to_numeric(df['First Tooltip'])
```

```python
df.info()
```

```python
df.nunique()
```

```python
df2 = df[df['First Tooltip'].notna()]
df2.nunique()
```

```python
df2 = df.dropna(how='any')
```

```python
df2.nunique()
```

```python
df3 = df2.pivot_table(columns='Indicator', index='Location', values='First Tooltip').reset_index()
```

```python
df4 = df3.dropna(axis = 0, how='any')
    
df4.info()
```

```python
df4.to_csv('WHO Data for PCA.csv', index=False)
```
